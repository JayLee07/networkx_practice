# Network Robustness

### The ability of maintain its general structures when there is an attack(Removing V or E) to a graph
### Robust network : Large min_node_cut and min_edge_cut

```python
G1 = nx.read_adjlist('dataset/karate.adjlist')
```

```python
## To disconnect G1, we only have to remove node '1'
print(nx.minimum_node_cut(G1))
```

```python
## To disconnect G1, we only have to remove edge e=(12,1)
print(nx.minimum_edge_cut(G1))
```
