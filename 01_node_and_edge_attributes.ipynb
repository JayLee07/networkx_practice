```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Show Edge attributes in undirected graph

```python
G = nx.Graph()
G.add_edge('A','B',weight=2.0,relation='family')
G.add_edge('C','D',weight=1.5,relation='family')
G.add_edge('A','C',weight=0.5,relation='enemy')
G.add_edge('B','D',weight=1.5,relation='friend')
nx.draw_networkx(G)
```

```python
print(G.edges()) #List of all edges
print(G.edges(data=True)) # List of all edges with attributes
print(G.edges(data='relation')) # LIst of all edges with relation=family
```

```python
G['B']['A'] # In undirected graph, edge order is not important
```

### Show edged attributes in directed graph

```python
G = nx.DiGraph()
G.add_edge('A','B',weight=2.0,relation='family')
G.add_edge('C','D',weight=1.5,relation='family')
G.add_edge('A','C',weight=0.5,relation='enemy')
G.add_edge('B','D',weight=1.5,relation='friend')
nx.draw_networkx(G)
```

```python
print(G.edges()) #List of all edges
print(G.edges(data=True)) # List of all edges with attributes
print(G.edges(data='relation')) # LIst of all edges with relation=family
```

```python
G['B']['A'] #In directed graph, edge order is important
```

### Show edge attributes in multigrph

```python
G = nx.MultiDiGraph()
G.add_edge('A','B',weight=2.0,relation='family')
G.add_edge('A','B',weight=1.0,relation='teacher-student')
G.add_edge('C','D',weight=1.5,relation='family')
G.add_edge('A','C',weight=0.5,relation='enemy')
G.add_edge('B','D',weight=1.5,relation='friend')
nx.draw_networkx(G)
```

```python
print(G.edges()) #List of all edges
print(G.edges(data=True)) # List of all edges with attributes
print(G.edges(data='relation')) # LIst of all edges with relation=family
```

```python
print(G['A']['B'][0])
print(G['A']['B'][1]) # Can find each relationship by numbering
```

### Show node attributes

```python
G = nx.MultiDiGraph()
G.add_edge('A','B',weight=2.0,relation='family')
G.add_edge('A','B',weight=1.0,relation='teacher-student')
G.add_edge('C','D',weight=1.5,relation='family')
G.add_edge('A','C',weight=0.5,relation='enemy')
G.add_edge('B','D',weight=1.5,relation='friend')

G.add_node('A', job='teacher')
G.add_node('A', role='mother')
G.add_node('B', job='student')
G.add_node('B', role='son')
G.add_node('C', job='parents')
G.add_node('D', job='parents')

nx.draw_networkx(G)
```

```python
print(G.nodes())
print(G.nodes(data=True)) # List of all edges with attributes
print(G.nodes(data='job')) # LIst of all edges with relation=family
```
