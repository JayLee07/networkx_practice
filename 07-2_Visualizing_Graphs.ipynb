```python
import pickle
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist('dataset/Employee_Movie_Choices.txt', delimiter='\t')
nx.draw_networkx(G)
```

```python
## Check what layouts are possible in networkX
[x for x in nx.__dir__() if x.endswith('_layout')]
```

```python
## Draw graph with random layout
plt.figure(figsize=(7,7))
pos = nx.random_layout(G)
nx.draw_networkx(G, pos=pos, fontsize=4)
```

```python
### Draw graph with circular layout
plt.figure(figsize=(10,10))
pos = nx.circular_layout(G)
nx.draw_networkx(G, pos=pos)
```

```python
# Draw the graph adding alpha, removing labels, and softening edge color
plt.figure(figsize=(8,6))
nx.draw_networkx(G, pos, alpha=0.7, with_labels=False, edge_color='.4')
plt.axis('off')
plt.tight_layout()
```
