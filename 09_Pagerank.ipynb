```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
undirected_G = nx.read_adjlist('dataset/karate.adjlist', create_using=nx.Graph(), nodetype=str)
directed_G = nx.read_edgelist('dataset/email_network.txt', delimiter='\t', \
                              create_using=nx.DiGraph(), nodetype=str,data = [('time', int)])
```

```python
pgrank1 = nx.pagerank(undirected_G, alpha=0.0) # Pagerank with no damping param
pgrank2_wrong = nx.pagerank(directed_G, alpha=0.0) # Without damping param, stucked in sink nodes --> Same value!
pgrank2 = nx.pagerank(directed_G)
```
