```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib inline
```

## Path: A seq of nodes connected by an edge
## Path length: # of steps it contains from beginning to end

```python
G = nx.read_adjlist('dataset/karate.adjlist')
nx.draw_networkx(G)
```

## Distance: The length of the shortest path between two nodes

```python
nx.shortest_path(G, source='1', target='27')
```

```python
nx.shortest_path_length(G, source='1', target='27')
```

## Find the distance from a node to every node --> Breadth-First serach

```python
T = nx.bfs_tree(G, source='1')
T.edges()
```

```python
nx.draw_networkx(T, font_size=8, node_size=150)
```

```python
nx.shortest_path_length(G, source='1')
```

## Distance Measure:  How to measure and characterize the all pairs of nodes in a graph?

### Approach1: Average distance
#### SImply average the distance of shortest path length

```python
nx.average_shortest_path_length(G)
```

### Approache2: Diameter
#### Maximum distance of among the pairs of nodes which are in a graph

```python
nx.diameter(G)
```

## Approach3: Radius
#### Eccentricity of a node A: the largest distance between A and all other nodes
#### Radius: Minimum eccentricity

```python
nx.eccentricity(G)
```

```python
nx.radius(G)
```

## Approch4: Periphery and Center
#### Periphery: the set of nodes that have same eccentricity to diameter (Maximum distance)
#### Center: the set of nodes that have same eccentricity to radius(Minimum distance)

```python
nx.periphery(G)
```

```python
nx.center(G)
```
