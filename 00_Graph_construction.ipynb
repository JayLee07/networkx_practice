```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Undirected Graph

```python
G=nx.Graph()
G.add_edge('A','B')
G.add_edge('C','D')
G.add_node('E')
nx.draw_networkx(G)
```

### Directed Graph

```python
G = nx.DiGraph()
G.add_edge('A','B')
G.add_edge('C','D')
G.add_node('E')
nx.draw_networkx(G)
```

### Weighted Undirected Graph

```python
G = nx.Graph()
G.add_edge('A','B',weight=2) # Two nodes become closer as weight increase
G.add_edge('C','D',weight=0.3)
G.add_node('E')
nx.draw_networkx(G)
```

### Weighted Directed Graph

```python
G = nx.DiGraph()
G.add_edge('A','B',weight=2)
G.add_edge('C','D',weight=0.3)
G.add_node('E')
nx.draw_networkx(G)
```

### Signed Graph

```python
G = nx.DiGraph()
G.add_edge('A','B',weight=3, sign='+')
G.add_edge('C','D',weight=0.2, sign='-')
G.add_node('E')
nx.draw_networkx(G)
```

### Other edge attributes

```python
G = nx.Graph()
G.add_edge('A','B',relation='friend')
G.add_edge('C','D',relation='enemy')
G.add_edge('A','C',relation='friend')
G.add_edge('B','D',relation='neutral')
G.add_node('E')
nx.draw_networkx(G)
```

### MultiGraph

```python
G = nx.MultiGraph()
G.add_edge('A','B',relation='friend')
G.add_edge('C','D',relation='enemy')
G.add_edge('A','B',relation='neutral')
G.add_edge('C','D',relation='enemy')
G.add_edge('A','C',relation='friend')
G.add_node('E')
nx.draw_networkx(G)
```
