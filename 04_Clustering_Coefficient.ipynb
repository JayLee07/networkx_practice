```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib inline
```

## Clustering coeff
#### measure the degree to which nodes in a network tends to 'cluster' or form triangles

### Local Clustering Coefficient
##### (# of pairs of a node's friends who are friends) / (# of pairs of a node's friends)

```python
G = nx.read_adjlist('dataset/karate.adjlist', nodetype=int)
```

```python
nx.draw_networkx(G)
```

```python
print(nx.clustering(G,1))
print(nx.clustering(G,2))
```

### Global Clustering Coeeficient
##### Approach 1. Average all local clustering coeff

```python
print(nx.average_clustering(G))
```

##### Approach 2. Transitivity
###### Open Triads: three nodes that are connected by two edges
###### A triangle has three different open triads
###### Transitivity = 3*(# of closed triads) / (# of open triads)

```python
print(nx.transitivity(G))
```

##### Both approaches measures the tendency of forming triangle
##### Transitivity weights nodes with large degree higher
