```python
from networkx.algorithms import bipartite
```

# Bipartite graph

#### A graph whose nodes can be split into two sets L and R. Every edge connects a node in L with a node in R

```python
G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E'], bipartite=0)
G.add_nodes_from([1,2,3,4], bipartite=1)
G.add_edges_from([('A',1),('B',1),('C',1),('C',3),('C',2),('D',2),('E',3),('E',4)])
nx.draw_networkx(G)
```

```python
# Check whether a graph is bipartite
print(bipartite.is_bipartite(G))

#Break the bipartite structure
G.add_edge('A','B')
print(bipartite.is_bipartite(G))
```

```python
G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E'], bipartite=0)
G.add_nodes_from([1,2,3,4], bipartite=1)
G.add_edges_from([('A',1),('B',1),('C',1),('C',3),('C',2),('D',2),('E',3),('E',4)])
```

```python
# Check a set of nodes is bipartite
print(bipartite.is_bipartite_node_set(G, ['A','B','C','D','E']))
print(bipartite.is_bipartite_node_set(G, [1,2,3,4]))
print(bipartite.is_bipartite_node_set(G, ['A','B','C','D','E',1]))
```

```python
# Getting each set of nodes of a bipartite graph
bipartite.sets(G)
```

# Projected Graphs

#### Network of nodes in group L, where a pair of nodes is connected if they have a common neighbor in R

```python
G = nx.Graph()
G.add_edges_from([('A',1),('B',1),('C',1),('D',1),('H',1),('B',2),('C',2),('D',2),('E',2),\
                ('G',2),('E',3),('F',3),('H',3),('J',3),('E',4),('I',4),('J',4)])
```

```python
X1 = set(['A','B','C','D','E','F','G','H','I','J'])
G1 = bipartite.projected_graph(G, X1)
```

```python
X2 = set([1,2,3,4])
G2 = bipartite.projected_graph(G, X2)
```

```python
X3 = set([1,2,3,4])
G3 = bipartite.weighted_projected_graph(G, X3)
```

```python
nx.draw_networkx(G)
```

```python
nx.draw_networkx(G1)
```

```python
nx.draw_networkx(G2)
```
